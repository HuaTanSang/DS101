## Vẽ đồ thị hàm mật độ sác suất của một số phân phối

```python
import matplotlib.pyplot as plt
import math
import numpy as np
```

1. Phân phối nhị thức

```python
def binomial_distribution(n, k, p):
    # Tính hệ số nhị thức
    nCk = math.comb(n, k)

    # Tính xác suất
    probability = nCk * (p ** k) * ((1 - p) ** (n - k))

    return probability

def draw_binomial_distribution(n, p): # Vẽ đồ thị

    x_values = list(range(n + 1))
    y_values = [binomial_distribution(n, k, p) for k in x_values]

    plt.bar(x_values, y_values, align = 'center', color = 'pink', alpha = 0.9)
    plt.xlabel('Số kết quả có lợi mong muốn')
    plt.ylabel('Xác suát (P(X = k))')
    plt.title(f'Phân phối nhị thức: n = {n}, p = {p}')
    plt.show()

# Mo phỏng với số thử nghiệm n = 101 và xác suất thành công p = 0.75
N = 101
p = 0.75

draw_binomial_distribution(N, p)
```

2.Phân phối Gauss

```python
def gaussian_distribution(x, mean, std_dev):
    exponent = -((x - mean) ** 2) / (2 * (std_dev ** 2))
    probability = (1 / (np.sqrt(2 * np.pi) * std_dev)) * np.exp(exponent)
    return probability

def draw_gaussian_distribution(mean, std_dev):
    x_values = np.linspace(mean - 4 * std_dev, mean + 4 * std_dev, 100)
    y_values = [gaussian_distribution(x, mean, std_dev) for x in x_values]

    plt.plot(x_values, y_values, label = f'N({mean}, {std_dev**2})')
    plt.xlabel('Giá trị (x)')
    plt.ylabel('Xác suất')
    plt.title('Phân phối Gaussian')
    plt.legend()
    plt.show()


# Thử nghiệm với giá trị trung bình mean = 5 và độ lệch chuẩn std_dev = 0.7
mean = 5
std_dev = 0.7

draw_gaussian_distribution(mean, std_dev)
```

3.Phân phối Poisson


```python

def poisson_distribution(k, lambd):
    probability = (np.exp(-lambd) * (lambd ** k)) / np.math.factorial(k)
    return probability

def plot_poisson_distribution(lambd, max_k):
    x_values = np.arange(0, max_k + 1, 1)
    y_values = [poisson_distribution(k, lambd) for k in x_values]

    plt.bar(x_values, y_values, align = 'center', color = 'pink', alpha = 0.9)
    plt.xlabel('Số lần sự kiện (k)')
    plt.ylabel('Xác suất')
    plt.title(f'Phân phối Poisson (λ = {lambd})')
    plt.show()

# Thử nghiệm với tỉ lệ sự kiện (kỳ vọng) là 3 và vẽ đồ thị cho số lần sự kiện tối đa là 10
lambd_value = 3
max_k_value = 10

plot_poisson_distribution(lambd_value, max_k_value)
```

4. Phân phối Bernulli

```python
def bernoulli_distribution(k, p):
    probability = p if k == 1 else 1 - p
    return probability

def draw_bernoulli_distribution(p):
    x_values = [0, 1]
    y_values = [bernoulli_distribution(k, p) for k in x_values]

    plt.bar(x_values, y_values, align = 'center', color = 'pink', alpha = 0.9)
    plt.xticks(x_values, ['Thất bại (0)', 'Thành công (1)'])
    plt.xlabel('Kết quả')
    plt.ylabel('Xác suất')
    plt.title(f'Phân phối Bernoulli (p = {p})')
    plt.show()

# Thử nghiệm với xác suất thành công p = 0.3
p = 0.3
draw_bernoulli_distribution(p)
```

5. Phân phối đều

```python
def uniform_distribution(x, a, b):
    if a <= x <= b:
        probability = 1 / (b - a)
    else:
        probability = 0
    return probability

def draw_uniform_distribution(a, b):

    x_values = np.linspace(a - 1, b + 1, 1000)
    y_values = [uniform_distribution(x, a, b) for x in x_values]

    plt.plot(x_values, y_values, label=f'U({a}, {b})')
    plt.xlabel('Giá trị x')
    plt.ylabel('Xác suất')
    plt.title('Phân phối đều')
    plt.legend()
    plt.show()

# Thử nghiệm với giá trị tối thiểu a = 1 và giá trị tối đa b = 10
min_value = 1
max_value = 10

draw_uniform_distribution(min_value, max_value)
```

6. Phân phối mũ

```python
def exponential_distribution(x, lambd):
    if x < 0:
        return 0
    probability = lambd * np.exp(-lambd * x)
    return probability

def draw_exponential_distribution(lambd):
    x_values = np.linspace(0, 5, 1000)
    y_values = [exponential_distribution(x, lambd) for x in x_values]

    plt.plot(x_values, y_values, label=f'Exp({lambd})')
    plt.xlabel('Giá trị x')
    plt.ylabel('Xác suất')
    plt.title('Phân phối mũ')
    plt.legend()
    plt.show()

# Thử nghiệm với tỷ lệ sự kiện (lambd) là 1
lambd = 1
draw_exponential_distribution(lambd)
```
