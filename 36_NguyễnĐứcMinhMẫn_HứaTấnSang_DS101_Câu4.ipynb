```python
import pandas as pd
import numpy as np
from collections import defaultdict
```

```python
df = pd.read_csv(r'M:\c++\breakfast.csv')
data = df.Food.tolist()
data
```

```python
# food_count: đếm số lần xuất hiện của một trạng thái
# food_pair_count: đếm tất cả các cặp trạng thái có thể (các món ăn hôm nay và ngay mai) [current][future]
food_count = defaultdict(int)
food_pair_count = defaultdict(lambda: defaultdict(float))
```

```python
# Đểm số thức ăn và xây dựng dữ liệu theo từng loại thức ăn tương ứng
for i in range(len(data)):
    food_count[data[i]] += 1
    if i == len(data) - 1:
        # self loop
        food_pair_count[data[i]][data[i]] += 1
        break
    food_pair_count[data[i]][data[i + 1]] += 1
```

```python
# chuẩn hóa xác suất theo t hàng
for key, value in food_pair_count.items():
    for k, v in value.items():
        food_pair_count[key][k] /= food_count[key] # chuẩn hóa
```

```python
# lấy index của các món ăn để dễ thao tác
keys = list(food_count.keys())
idx = range(len(keys))
key_to_idx = dict(zip(keys, idx)) # key to index
print(key_to_idx)
```

```python
# Xây dựng ma trận dựa vào xác suất đã chuẩn hóa và index ở trên
P = []
for key, value in food_pair_count.items():
    P.append(list(value.values()))

# chuyển list sang numpy để dễ tính toán
P = np.array(P)

print('Ma trận chuyển trạng thái P: ')
print(P)
```

```python
# Kiểm tra tổng hàng bằng 1
if(sum(P[0]) + sum(P[1]) + sum(P[2]) + sum(P[3])) == 4:
    print("Ma trận phù hợp")
```

```python
# dự đoán món ăn
curr_food = data[-1] #Lấy ra món ăn cuối cùng
curr_distribution = P[key_to_idx[curr_food]]
predicted_food = np.random.choice(keys, p=curr_distribution) # random walk with known distribution
predicted_probability = P[key_to_idx[curr_food]][key_to_idx[predicted_food]] #Tỉ lệ xảy ra
```

```python
print(f'Món ăn chúng ta ăn hôm trước: {data[-1]}')
print(f'Món ăn nên ăn vào hôm nay là "{predicted_food}" với khả năng xảy ra là {round(predicted_probability * 100, 2)}%')
```
