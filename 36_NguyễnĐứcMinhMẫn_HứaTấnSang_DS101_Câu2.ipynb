### Viết hàm để thực hiện thống kê mô tả

Lấy dữ liệu từ cột Final Price để xử lý


```python
d = df["Final Price"].to_numpy()
```

Viết hàm thống kê

```python
def des(data):
    KQ = {}
    def count(data):
        return data.shape[0]
    def mean (data):
        return np.sum(data)/len(data)
    def var(data):
        re = 0
        for i in range (len(data)):
            re += (data[i] - mean(data))**2
        return re/(len(data) - 1)
    def std(data):
        return (var(data))**0.5
    def min(data):
        return np.min(data)
    def max(data):
        return np.max(data)
    def median(data):
        m = np.sort(data)
        if len(data) % 2 == 0:
            return (m[(int(len(m) / 2))] + m[(int((len(m)+2) / 2))]) / 2
        else:
            return m[(int((len(m)+1) / 2))]
    def mode(data):
        distinct_vals, cnt=np.unique(data,return_counts=True)
        return distinct_vals[np.argmax(cnt)]
    KQ['count']=count(data)
    KQ['mean']=mean(data)
    KQ['var']=var(data)
    KQ['std']=std(data)
    KQ['min']=min(data)
    KQ['max']=max(data)
    KQ['median']=median(data)
    KQ['mode']=mode(data)
    return KQ

```
